# Matplotlib; Adding custom Axes

*This notebook is part of an article. It act like a playground for the instructions and information present in the article. View this article [here](https://medium.com/@HanzalaAhmed/pyplot-add-axes-in-matplotlib-a-challenging-encounter-fded232c31bf).*

```python
from matplotlib import pyplot as plt
```

### *Figure with two Axes; missing proper Dimensions*

```python
# Initialize Figure
figure = plt.figure( figsize = ( 6, 4 ) )

# Add axes
ax1 = figure.add_axes( [ 0, 0, 2, 2 ] )
ax2 = figure.add_axes( [ 0, 0, 2, 2 ] )

# Plot simple line graphs
ax1.plot( [0.4, 0.6], [0.6, 0.4] )
ax2.plot( [0.4, 0.6], [0.4, 0.6] )

# Render Plots
plt.show()
```

### *Figure with two Axes; Overlap*

```python
# Initialize Figure
figure = plt.figure( figsize = ( 6, 4 ) )

# Add axes
ax1 = figure.add_axes( [ 0, 0, 2/6, 2/4 ] )
ax2 = figure.add_axes( [ 0, 0, 2/6, 2/4 ] )

# Plot simple line graphs
ax1.plot( [0.4, 0.6], [0.6, 0.4] )
ax2.plot( [0.4, 0.6], [0.4, 0.6] )

# Render Plots
plt.show()
```

### *Figure with two Axes; Adjacent*

```python
# Initialize Figure
figure = plt.figure( figsize = ( 6, 4 ) )

# Add axes
ax1 = figure.add_axes( [ 0, 0, 2/6, 2/4 ] )
ax2 = figure.add_axes( [ 2/6, 0, 2/6, 2/4 ] )

# Plot simple line graphs
ax1.plot( [0.4, 0.6], [0.6, 0.4] )
ax2.plot( [0.4, 0.6], [0.4, 0.6] )

# Render Plots
plt.show()
```

### *Figure with two Axes; Horizontal Alingment*

```python
# Initialize Figure
figure = plt.figure( figsize = ( 6, 4 ) )

# Add axes
ax1 = figure.add_axes( [ 0, 0, 2/6, 2/4 ] )
ax2 = figure.add_axes( [ 3/6, 0, 2/6, 2/4 ] )

# Plot simple line graphs
ax1.plot( [0.4, 0.6], [0.6, 0.4] )
ax2.plot( [0.4, 0.6], [0.4, 0.6] )

# Render Plots
plt.show()
```

### *Figure with two Axes; Vertical Alingment*

```python
# Initialize Figure
figure = plt.figure( figsize = ( 6, 4 ) )

# Add axes
ax1 = figure.add_axes( [ 0, 2.5/4, 2/6, 2/4 ] )
ax2 = figure.add_axes( [ 0, 0, 2/6, 2/4 ] )

# Plot simple line graphs
ax1.plot( [0.4, 0.6], [0.6, 0.4] )
ax2.plot( [0.4, 0.6], [0.4, 0.6] )

# Render Plots
plt.show()
```
